```python
import networkx as nx
import matplotlib.pyplot as plt
import itertools


# Define the input graphs - G & H
G = nx.cycle_graph(3)
H = nx.path_graph(2)


# Calculate the number of nodes of input graphs - G & H
GV = G.nodes()
HV = H.nodes()
J = len(GV)
K = len(HV)


# Define & calculate the Cartesian Product of the nodes of input graphs - G & H
X = []
for e in itertools.product(GV,HV):
    X.append(e)
sorted(X)
L = len(X)


# Define a representational list of nodes Y that is of same length as X. Y will be used for plotting Cartesian Product of the graph
Y = [*range(0,L,1)]


# Check the criteria for adjacency formation in the Cartesian Product of the graph and create a list of edge E
E = []
p = 1
for i in range(L):
    for j in range(p,L,1):
        if (X[i][0]==X[j%L][0] and H.has_edge(X[i][1],X[j%L][1])) or (X[i][1]==X[j%L][1] and G.has_edge(X[i][0],X[j%L][0])):
            E.append((Y[i],Y[j%L]))
    p= L%j + p

        
# Define and create the representational Cartesian Product of the graph with Y as list of nodes and E as list of edges
I = nx.Graph()
I.add_nodes_from(Y)
I.add_edges_from(E)



# Plot the input and output graphs
print("This is how the first graph G looks like:\n")
plt.figure(2, figsize=(2,2))
nx.draw(G, with_labels=1)
plt.show()


print("This is how the second graph H looks like:\n")
plt.figure(2, figsize=(2,2))
nx.draw(H, with_labels=1)
plt.show()


print("This is how the cartesian product of the graphs GxH looks like:\n")
plt.figure(2, figsize=(3,3))
nx.draw(I, with_labels=1)
plt.show()

# Output the node mappings between Y and X
print("The node mappings of between Y i.e. representational and X i.e. actual is:\n")
print(dict(zip(Y,X)))

```
